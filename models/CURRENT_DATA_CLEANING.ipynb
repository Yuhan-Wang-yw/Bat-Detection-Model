```python
# Add Fungus Phylum, Class, Order, Family, Genus, and Species for df_white_nose_county_status
# Fungus value = WNS Fungus value if determination == WNS Positive or Pd Positive, else = NaN
# WNS Fungus-- Phylum: 'Ascomycota', Class: 'Leotiomycetes', Order: 'Thelebolales',
#               Family: 'Pseudeurotiaceae', Genus: 'Pseudogymnoascus', Species: 'destructans'

fungus_phylum = {'WNS Positive' : 'Ascomycota', 'Pd Positive' : 'Ascomycota', 'WNS Suspect': np.nan, 'Pd Presumed': np.nan}
fungus_class = {'WNS Positive' : 'Leotiomycetes','Pd Positive' : 'Leotiomycetes', 'WNS Suspect': np.nan, 'Pd Presumed': np.nan}
fungus_order = { 'WNS Positive': 'Thelebolales','Pd Positive': 'Thelebolales', 'WNS Suspect': np.nan, 'Pd Presumed': np.nan}
fungus_family = {'WNS Positive': 'Pseudeurotiaceae','Pd Positive': 'Pseudeurotiaceae', 'WNS Suspect': np.nan, 'Pd Presumed': np.nan}
fungus_genus = { 'WNS Positive': 'Pseudogymnoascus','Pd Positive': 'Pseudogymnoascus', 'WNS Suspect': np.nan, 'Pd Presumed': np.nan}
fungus_species = {'WNS Positive': 'destructans','Pd Positive': 'destructans', 'WNS Suspect': np.nan, 'Pd Presumed': np.nan}

df_white_nose_county_status['Fungus Classification - Phylum'] = df_white_nose_county_status['determination'].map(fungus_phylum)
df_white_nose_county_status['Fungus Classification - Class'] = df_white_nose_county_status['determination'].map(fungus_class)
df_white_nose_county_status['Fungus Classification - Order'] = df_white_nose_county_status['determination'].map(fungus_order)
df_white_nose_county_status['Fungus Classification - Family'] = df_white_nose_county_status['determination'].map(fungus_family)
df_white_nose_county_status['Fungus Classification - Genus'] = df_white_nose_county_status['determination'].map(fungus_genus)
df_white_nose_county_status['Fungus Classification - Species'] = df_white_nose_county_status['determination'].map(fungus_species)

df_white_nose_county_status.head()
```
